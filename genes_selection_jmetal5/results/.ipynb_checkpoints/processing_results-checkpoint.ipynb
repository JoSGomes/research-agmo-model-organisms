```python
from scipy.stats import wilcoxon
from scipy.io import arff
import pandas as pd
import numpy as np
import os
```

```python
dirs_classifiers = os.listdir()
dirs_classifiers.pop(0) # Removing dir .ipynb_checkpoints, if you haven't that directory, remove this code line.
all_results = {}
all_results_without_selection = {}
for classifier in dirs_classifiers:
    path = ""
    if not os.path.isfile(classifier):
        path += classifier + "/"
        dirs_datasets = os.listdir(path)
        for dataset in dirs_datasets:
                if not classifier in all_results.keys():
                    all_results[classifier] = {}
                    all_results_without_selection[classifier] = {}
                    print(path + dataset + "/" + "results.csv") 
                all_results[classifier][dataset] = pd.read_csv(path + dataset + "/" + "results.csv")
                all_results_without_selection[classifier][dataset] = pd.read_csv(path + dataset + "/" + "results-without-selection.csv")
```

```python
muxColumns = pd.MultiIndex.from_product([['Sem seleção de características', 'Com seleção de características'], ['J48','KNN', 'NB']])
index = pd.MultiIndex.from_product([['D. melanogaster','M. musculus','C. elegans', 'S. cerevisiae'], ['BP', 'MF', 'CC', 'BP.MF', 'BP.CC', 'MF.CC', 'BP.MF.CC']])

muxColumnsTSC = pd.MultiIndex.from_product([['Taxa de seleção de características (TSC)'], ['J48','KNN', 'NB']])
indexTSC = pd.MultiIndex.from_product([['D. melanogaster','M. musculus','C. elegans', 'S. cerevisiae'], ['BP', 'MF', 'CC', 'BP.MF', 'BP.CC', 'MF.CC', 'BP.MF.CC']])

muxColumns_wilcoxon = pd.MultiIndex.from_product([['Valor de P'], ['J48','KNN', 'NB']])
index_wilcoxon = pd.MultiIndex.from_product([['D. melanogaster','M. musculus','C. elegans', 'S. cerevisiae'], ['BP', 'MF', 'CC', 'BP.MF', 'BP.CC', 'MF.CC', 'BP.MF.CC']])
```

```python
table_results = pd.DataFrame(index=index, columns=muxColumns, dtype='string')
table_results_tsc = pd.DataFrame(index=indexTSC, columns=muxColumnsTSC, dtype='string')
table_wilcoxon_test = pd.DataFrame(index=index_wilcoxon, columns=muxColumns_wilcoxon, dtype='string')
```

```python
classifiers = ['J48', 'KNN', 'NB']
datasets = ['BP', 'MF', 'CC', 'BPMF', 'BPCC', 'MFCC', 'BPMFCC']
organisms = ['Drosophila melanogaster','Mus musculus','Caenorhabditis elegans', 'Saccharomyces cerevisiae']

for classifier in classifiers:
    for dataset in datasets:
            fly = all_results[classifier][dataset][(all_results[classifier][dataset]["organism"] == "Drosophila melanogaster")]
            mouse = all_results[classifier][dataset][(all_results[classifier][dataset]["organism"] == "Mus musculus")]
            worm = all_results[classifier][dataset][(all_results[classifier][dataset]["organism"] == "Caenorhabditis elegans")]
            yeast = all_results[classifier][dataset][(all_results[classifier][dataset]["organism"] == "Saccharomyces cerevisiae")]

            fly_without_selection = all_results_without_selection[classifier][dataset][(all_results_without_selection[classifier][dataset]["organism"] == "Drosophila melanogaster")]
            mouse_without_selection = all_results_without_selection[classifier][dataset][(all_results_without_selection[classifier][dataset]["organism"] == "Mus musculus")]
            worm_without_selection = all_results_without_selection[classifier][dataset][(all_results_without_selection[classifier][dataset]["organism"] == "Caenorhabditis elegans")]
            yeast_without_selection = all_results_without_selection[classifier][dataset][(all_results_without_selection[classifier][dataset]["organism"] == "Saccharomyces cerevisiae")]

            mean_gmean_fly = '{0:.2f}'.format((np.mean(fly["GMean"])*100))
            std_gmean_fly = '{0:.2f}'.format((np.std(fly["GMean"])*100))

            mean_gmean_mouse = '{0:.2f}'.format((np.mean(mouse["GMean"])*100))
            std_gmean_mouse = '{0:.2f}'.format((np.std(mouse["GMean"])*100))
            
            mean_gmean_worm = '{0:.2f}'.format((np.mean(worm["GMean"])*100))
            std_gmean_worm = '{0:.2f}'.format((np.std(worm["GMean"])*100))

            mean_gmean_yeast = '{0:.2f}'.format((np.mean(yeast["GMean"])*100))
            std_gmean_yeast = '{0:.2f}'.format((np.std(yeast["GMean"])*100))


            mean_gmean_fly_without_selection = '{0:.2f}'.format((np.mean(fly_without_selection["GMean"])*100))
            std_gmean_fly_without_selection = '{0:.2f}'.format((np.std(fly_without_selection["GMean"])*100))
        
            mean_gmean_mouse_without_selection = '{0:.2f}'.format((np.mean(mouse_without_selection["GMean"])*100))
            std_gmean_mouse_without_selection = '{0:.2f}'.format((np.std(mouse_without_selection["GMean"])*100))
        
            mean_gmean_worm_without_selection = '{0:.2f}'.format((np.mean(worm_without_selection["GMean"])*100))
            std_gmean_worm_without_selection = '{0:.2f}'.format((np.std(worm_without_selection["GMean"])*100))
        
            mean_gmean_yeast_without_selection = '{0:.2f}'.format((np.mean(yeast_without_selection["GMean"])*100))
            std_gmean_yeast_without_selection = '{0:.2f}'.format((np.std(yeast_without_selection["GMean"])*100))

            datasetTable = dataset
            if datasetTable == 'BPMF':
                datasetTable = "BP.MF"
            elif datasetTable == 'BPCC':
                datasetTable = "BP.CC"
            elif datasetTable == 'MFCC':
                datasetTable = "MF.CC"
            elif datasetTable == 'BPMFCC':
                datasetTable = "BP.MF.CC"
             
            try:
                table_wilcoxon_test.loc[('D. melanogaster', datasetTable), ('Valor de P', classifier)] = str((wilcoxon(fly["GMean"], fly_without_selection["GMean"]).pvalue)).replace(".", ",")
            except RuntimeError as error:
                table_wilcoxon_test.loc[('D. melanogaster', datasetTable), ('Valor de P', classifier)] = str(0)
                print(error + f": Fly / {classifier} / {datasetTable}")
            try:
                table_wilcoxon_test.loc[('M. musculus', datasetTable), ('Valor de P', classifier)] = str((wilcoxon(mouse["GMean"], mouse_without_selection["GMean"]).pvalue)).replace(".", ",")
            except RuntimeError as error:
                table_wilcoxon_test.loc[('M. musculus', datasetTable), ('Valor de P', classifier)] = str(0)
                print(error + f": Mouse / {classifier} / {datasetTable}")
            try:
                table_wilcoxon_test.loc[('C. elegans', datasetTable), ('Valor de P', classifier)] = str((wilcoxon(worm["GMean"], worm_without_selection["GMean"]).pvalue)).replace(".", ",")
            except RuntimeError as error:
                table_wilcoxon_test.loc[('C. elegans', datasetTable), ('Valor de P', classifier)] = str(0)
                print(error + f": Worm / {classifier} / {datasetTable}")
            if not (np.all(yeast["GMean"] == 0) or np.all(yeast_without_selection["GMean"] == 0) or np.all(yeast["GMean"] == yeast_without_selection["GMean"]) ):
                table_wilcoxon_test.loc[('S. cerevisiae', datasetTable), ('Valor de P', classifier)] = str((wilcoxon(yeast["GMean"], yeast_without_selection["GMean"]).pvalue)).replace(".", ",")
            else:
                table_wilcoxon_test.loc[('S. cerevisiae', datasetTable), ('Valor de P', classifier)] = str(0)
        
                
        
            table_results.loc[('D. melanogaster', datasetTable), ('Com seleção de características', classifier)] = (mean_gmean_fly + ' ± ' + std_gmean_fly).replace(".", ",")
            table_results.loc[('M. musculus', datasetTable), ('Com seleção de características', classifier)] = (mean_gmean_mouse + ' ± ' + std_gmean_mouse).replace(".", ",")
            table_results.loc[('C. elegans', datasetTable), ('Com seleção de características', classifier)] = (mean_gmean_worm + ' ± ' + std_gmean_worm).replace(".", ",")
            table_results.loc[('S. cerevisiae', datasetTable), ('Com seleção de características', classifier)] = (mean_gmean_yeast + ' ± ' + std_gmean_yeast).replace(".", ",")

            table_results.loc[('D. melanogaster', datasetTable), ('Sem seleção de características', classifier)] = (mean_gmean_fly_without_selection + ' ± ' + std_gmean_fly_without_selection).replace(".", ",")
            table_results.loc[('M. musculus', datasetTable), ('Sem seleção de características', classifier)] = (mean_gmean_mouse_without_selection + ' ± ' + std_gmean_mouse_without_selection).replace(".", ",")
            table_results.loc[('C. elegans', datasetTable), ('Sem seleção de características', classifier)] = (mean_gmean_worm_without_selection + ' ± ' + std_gmean_worm_without_selection).replace(".", ",")
            table_results.loc[('S. cerevisiae', datasetTable), ('Sem seleção de características', classifier)] = (mean_gmean_yeast_without_selection + ' ± ' + std_gmean_yeast_without_selection).replace(".", ",")

    
```

```python
datasets = ['BP', 'MF', 'CC', 'BPMF', 'BPCC', 'MFCC', 'BPMFCC']
organisms = ['Drosophila melanogaster','Mus musculus','Caenorhabditis elegans', 'Saccharomyces cerevisiae']

for dataset in datasets:
        fly_1 = all_results["KNN"][dataset][(all_results["KNN"][dataset]["organism"] == "Drosophila melanogaster")]
        mouse_1 = all_results["KNN"][dataset][(all_results["KNN"][dataset]["organism"] == "Mus musculus")]
        worm_1 = all_results["KNN"][dataset][(all_results["KNN"][dataset]["organism"] == "Caenorhabditis elegans")]
        yeast_1 = all_results["KNN"][dataset][(all_results["KNN"][dataset]["organism"] == "Saccharomyces cerevisiae")]

        fly_2 = all_results["NB"][dataset][(all_results["NB"][dataset]["organism"] == "Drosophila melanogaster")]
        mouse_2 = all_results["NB"][dataset][(all_results["NB"][dataset]["organism"] == "Mus musculus")]
        worm_2 = all_results["NB"][dataset][(all_results["NB"][dataset]["organism"] == "Caenorhabditis elegans")]
        yeast_2 = all_results["NB"][dataset][(all_results["NB"][dataset]["organism"] == "Saccharomyces cerevisiae")]
        
        datasetTable = dataset
        if datasetTable == 'BPMF':
            datasetTable = "BP.MF"
        elif datasetTable == 'BPCC':
            datasetTable = "BP.CC"
        elif datasetTable == 'MFCC':
            datasetTable = "MF.CC"
        elif datasetTable == 'BPMFCC':
            datasetTable = "BP.MF.CC"
         
        try:
            table_KNNNB_wilcoxon_test.loc[('D. melanogaster', datasetTable), ('Valor de P KNN/NB')] = str((wilcoxon(fly_1["GMean"], fly_2["GMean"]).pvalue)).replace(".", ",")
        except RuntimeError as error:
            table_KNNNB_wilcoxon_test.loc[('D. melanogaster', datasetTable), ('Valor de P KNN/NB')] = str(0)
        try:
            table_KNNNB_wilcoxon_test.loc[('M. musculus', datasetTable), ('Valor de P KNN/NB')] = str((wilcoxon(mouse_1["GMean"], mouse_2["GMean"]).pvalue)).replace(".", ",")
        except RuntimeError as error:
            table_KNNNB_wilcoxon_test.loc[('M. musculus', datasetTable), ('Valor de P KNN/NB')] = str(0)
        try:
            table_KNNNB_wilcoxon_test.loc[('C. elegans', datasetTable), ('Valor de P KNN/NB')] = str((wilcoxon(worm_1["GMean"], worm_2["GMean"]).pvalue)).replace(".", ",")
        except RuntimeError as error:
            table_KNNNB_wilcoxon_test.loc[('C. elegans', datasetTable), ('Valor de P KNN/NB')] = str(0)
        if not (np.all(yeast_1["GMean"] == 0) or np.all(yeast_2["GMean"] == 0) or np.all(yeast_1["GMean"] == yeast_2["GMean"]) ):
            table_KNNNB_wilcoxon_test.loc[('S. cerevisiae', datasetTable), ('Valor de P KNN/NB')] = str((wilcoxon(yeast_1["GMean"], yeast_2["GMean"]).pvalue)).replace(".", ",")
        else:
            table_KNNNB_wilcoxon_test.loc[('S. cerevisiae', datasetTable), ('Valor de P KNN/NB')] = str(0)
    
```
