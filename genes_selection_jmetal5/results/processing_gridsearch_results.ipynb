```python
import pandas as pd
```

```python
df = pd.read_csv("results-gridsearch.csv")
```

```python
df = df.sort_values(by=['GMean', 'selectionRate'], ascending=[False, True])
```

```python
df['organism'].value_counts()
```

```python
fly = df[(df['organism'] == 'Drosophila melanogaster')]
mouse = df[(df['organism'] == 'Mus musculus')]
worm = df[(df['organism'] == 'Caenorhabditis elegans')]
yeast = df[(df['organism'] == 'Saccharomyces cerevisiae')]
```

```python
df['runningDataset'].value_counts()
```

```python
fly_cc = fly[(fly['runningDataset'] == 'CC')]
fly_mf = fly[(fly['runningDataset'] == 'MF')]
fly_mfcc = fly[(fly['runningDataset'] == 'MFCC')]
fly_bp = fly[(fly['runningDataset'] == 'BP')]
fly_bpcc = fly[(fly['runningDataset'] == 'BPCC')]
fly_bpmf = fly[(fly['runningDataset'] == 'BPMF')]
fly_bpmfcc = fly[(fly['runningDataset'] == 'BPMFCC')]

mouse_cc = mouse[(mouse['runningDataset'] == 'CC')]
mouse_mf = mouse[(mouse['runningDataset'] == 'MF')]
mouse_mfcc = mouse[(mouse['runningDataset'] == 'MFCC')]
mouse_bp = mouse[(mouse['runningDataset'] == 'BP')]
mouse_bpcc = mouse[(mouse['runningDataset'] == 'BPCC')]
mouse_bpmf = mouse[(mouse['runningDataset'] == 'BPMF')]
mouse_bpmfcc = mouse[(mouse['runningDataset'] == 'BPMFCC')]

worm_cc = worm[(worm['runningDataset'] == 'CC')]
worm_mf = worm[(worm['runningDataset'] == 'MF')]
worm_mfcc = worm[(worm['runningDataset'] == 'MFCC')]
worm_bp = worm[(worm['runningDataset'] == 'BP')]
worm_bpcc = worm[(worm['runningDataset'] == 'BPCC')]
worm_bpmf = worm[(worm['runningDataset'] == 'BPMF')]
worm_bpmfcc = worm[(worm['runningDataset'] == 'BPMFCC')]

yeast_cc = yeast[(yeast['runningDataset'] == 'CC')]
yeast_mf = yeast[(yeast['runningDataset'] == 'MF')]
yeast_mfcc = yeast[(yeast['runningDataset'] == 'MFCC')]
yeast_bp = yeast[(yeast['runningDataset'] == 'BP')]
yeast_bpcc = yeast[(yeast['runningDataset'] == 'BPCC')]
yeast_bpmf = yeast[(yeast['runningDataset'] == 'BPMF')]
yeast_bpmfcc = yeast[(yeast['runningDataset'] == 'BPMFCC')]
```
